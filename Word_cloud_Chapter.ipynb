```python
import wordcloud
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data=open("/content/Joshua.txt",mode="r",encoding="UTF-8").read()
```

```python
data
```

```python
stopword=wordcloud.STOPWORDS
```

```python
print(stopword)
```

```python
word_cloud=wordcloud.WordCloud(collocations=False,stopwords=stopword,background_color="pink",height=1600,width=1600).generate(data)
```

```python
plt.imshow(word_cloud,interpolation='bilinear')

plt.axis('off')
plt.show()
```

```python
word_cloud.to_file('Joshua_chapter_2.png')
```
